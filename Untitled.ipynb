```python
%matplotlib inline

import json
from pprint import pprint
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = json.load(open('concrete_reward_accuracies.json'))
```

```python
#pprint(data)
x =[]
length = len(data["0"]["x"])
for i in range(length):
    x.append(data["0"]["x"][i])
print(x)
```

```python
mean = []
variance = []
minValue = []
maxValue = []
sd = []
length = len(data["0"]["y"])
for j in range(length):
    dataarray = []
    for i in data:
        #print(i)
        #print(j)
        dataarray.append(data[i]["y"][j])
    mean.append(np.mean(dataarray))
    variance.append(np.var(dataarray))
    maxValue.append(np.amax(dataarray))
    minValue.append(np.amin(dataarray))
    sd.append(np.std(dataarray))
    #print(dataarray)
print(minValue)
```

```python
stdUp = np.add(mean ,sd)
stdDown = np.subtract(mean ,sd)
```

```python
def getGraphData(data):
    x =[]
    x_length = len(data["0"]["x"])
    for i in range(x_length):
        x.append(data["0"]["x"][i])
    mean = []
    variance = []
    minValue = []
    maxValue = []
    sd = []
    length = len(data["0"]["y"])
    for j in range(length):
        dataarray = []
        for i in data:
            #print(i)
            #print(j)
            dataarray.append(data[i]["y"][j])
        mean.append(np.mean(dataarray))
        variance.append(np.var(dataarray))
        maxValue.append(np.amax(dataarray))
        minValue.append(np.amin(dataarray))
        sd.append(np.std(dataarray))
    
    stdUp = np.add(mean ,sd)
    stdDown = np.subtract(mean ,sd)
    
    return x,mean,variance,maxValue,minValue,sd,stdUp,stdDown
```
